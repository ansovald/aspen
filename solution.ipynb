```python
from call_clingo import generate_models
from model_handling import print_moves

for disks in range(3, 5):
    print(f"Generating models for {disks} disks:")
    model, time = generate_models(disks=disks)
    print_moves(model)
    print(f"Time taken: {time['total_time']:.4f} seconds for {disks} disks. Moves: {model['moves']}")
    print("\n\n\n")
```
